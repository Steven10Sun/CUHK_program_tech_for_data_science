```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
import warnings
warnings.filterwarnings('ignore')
```

```python
n = [i for i in range(1000, 5000000, 10000)]
pi_result = []
for i in n:
    x = np.random.uniform(0, 1, i)
    theta = np.random.uniform(0, 2*np.pi, i)
    
    # b.
    x_end1 = x - np.cos(theta)
    x_end2 = x + np.cos(theta)
    
    # c.
    min_end = np.minimum(x_end1, x_end2)
    max_end = np.maximum(x_end1, x_end2)
    
    # Condition: smaller < 0 and larger > 0
    condition = np.logical_and(min_end < 0, max_end > 0)
    
    p_cross_line = np.sum(condition)/i
    pi = 2/p_cross_line
    pi_result.append(pi)
    print(i)
```

```python
plt.plot(n, pi_result)
avg_pi = sum(pi_result)/100
# plt.axhline(avg_pi, color='r')
plt.show()
print(avg_pi)
```
